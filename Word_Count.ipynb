```python
raw_data1= amazon
for col in amazon.columns:
  raw_data1 = raw_data1.withColumnRenamed(col,col.replace(".", ""))
```

```python
raw_data1.select('reviewstext').show(2)
```

```python
from pyspark.sql import SparkSession

spark = SparkSession \
    .builder \
    .getOrCreate()

sc = spark.sparkContext
```

```python
df = raw_data1.filter(raw_data1.reviewstext.isNotNull())
```

```python
rdd1 = df.select('reviewstext').rdd.flatMap(lambda x:x)
```

```python
counts = rdd1.flatMap(lambda line: line.split(" ")).map(lambda word: (word, 1)).reduceByKey(lambda a, b: a + b)
```
